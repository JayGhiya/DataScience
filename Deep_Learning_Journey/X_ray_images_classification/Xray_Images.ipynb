# Formating dataset

```python
xray_df = pd.DataFrame(columns=["id","Class"])
```

```python
for x in xray_files:
    x_d = dict()
    x_d["id"] = x
    if x[len(x)-5] == '0':
        x_d['Class'] = "Normal"
    elif x[len(x)-5] == '1':
        x_d['Class'] = "Abnormal"
    
    xray_df = xray_df.append(pd.Series(data=x_d),ignore_index=True)

```

# Calculating mean

```python
mean = [0, 0, 0]
files = os.listdir(f'{PATH}CXR_png/')
x = np.array([np.array(Image.open(f'{PATH}CXR_png/' + fname)) for fname in train_xray_df['id']])
for image in x:
    mean += np.mean(image, axis=(0, 1))
mean = mean / len(x) / 255
```

```python
print(mean)
```

# Calculating Standard Deviation

```python
std = []

x = np.array([np.array(Image.open(f'{PATH}CXR_png/' + fname)) for fname in train_xray_df['id']])

for image in x:
    print(image)
    c = 0
    r_std = image[:,:,c].flatten()
    c+=1
    g_std = image[:,:,c].flatten()
    c+=1
    b_std = image[:,:,c].flatten()
    std.append(np.array([r_std,g_std,b_std]))
    
np_std = np.array(std)

r_channel = np_std[:,0].flatten()
g_channel = np_std[:,1].flatten()
b_channel = np_std[:,2].flatten()
r_norm = np.std(r_channel) / 255
g_norm = np.std(g_channel) / 255
b_norm = np.std(b_channel) / 255
imgs_std = np.array([r_norm,g_norm,b_norm])
print(imgs_std)
```
